```python
# Grouping by two categorical variables and aggregating the counts
grouped_data = data.groupby(['Gender', 'Specialization']).size().unstack()

# Plotting the stacked bar plot
grouped_data.plot(kind='bar', stacked=True)

# Adding labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Stacked Bar Plot of Specialization by Gender')

# Showing the plot
plt.show()
```

**Gender vs Specialization**

```python
# Relationship between Gender and Specialization
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Specialization', hue='Gender')
plt.title('Countplot of Specialization by Gender')
plt.xticks(rotation=45)
plt.show()
```

```python

# Step 5: Research Questions
# Test the claim about Computer Science Engineering salary
cs_eng_salary = data[data['Specialization'] == 'computer science']['Salary']
mean_salary = cs_eng_salary.mean()
print(f"Mean salary of Computer Science Engineers: {mean_salary} INR")
print("Claim is True" if 250000 <= mean_salary <= 300000 else "Claim is False")
```

```python
from scipy import stats
```

```python

# Relationship between gender and specialization
gender_specialization_cross = pd.crosstab(data['Gender'], data['Specialization'])
chi2, p, _, _ = stats.chi2_contingency(gender_specialization_cross)
print(f"P-value for Chi-square test: {p}")
if p < 0.05:
    print("There is a significant relationship between gender and specialization.")
else:
    print("There is no significant relationship between gender and specialization.")
```
